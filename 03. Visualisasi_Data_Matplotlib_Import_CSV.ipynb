##### Import CSV (Contoh Kedua: Dataset Spotify)

```python
import pandas as pd
pd = pd.read_csv("Dataset/dataspotify.csv")
pd
```

Task: Membuat plot bar atau grafik batang dengan memvisualisasikan Jumlah Pendengar per Bulan berdasarkan Nama Band menggunakan dataset tersebut.

```python
nama_band = list(pd[ 'Nama Band' ])
ppb = list(pd[ 'Pendengar per Bulan'])
print(nama_band)

print(ppb)
```

##### Plot Bar Version

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

nama_band = list(pd[ 'Nama Band' ])
ppb = list(pd[ 'Pendengar per Bulan'])
plt.barh(nama_band,ppb)

plt.title('Jumlah Pendengar per Bulan di Spotify')
plt.xlabel('Pendengar per Bulan')
plt.ylabel('Nama Band')

plt.show()
```

##### Plot Garis Version

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.style.use('fivethirtyeight')

nama_band = list(pd[ 'Nama Band' ])
ppb = list(pd[ 'Pendengar per Bulan'])
plt.plot(ppb,nama_band)

plt.title('Jumlah Pendengar per Bulan di Spotify')
plt.xlabel('Pendengar per Bulan')
plt.ylabel('Nama Band')

plt.show()
```
